### 2. How many cars have mpg ≥ 35?

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code here
# Filter cars with horsepower > 100 and weight < 3000
filtered_cars = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]

# Find the most common origin
most_common_origin = filtered_cars['origin'].mode().values[0]

print("Most common origin:", most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
#your code here
# Filter cars from Japan
japan_cars = df[df['origin'].str.lower() == 'japan']

# Calculate the mean acceleration
mean_acceleration = round(japan_cars['acceleration'].mean(), 2)

print("Mean acceleration of cars from Japan:", mean_acceleration)
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
# Step 1: Calculate median mpg
median_mpg = df['mpg'].median()

# Step 2: Filter cars with mpg above the median
above_median_mpg = df[df['mpg'] > median_mpg]

# Step 3: Make sure hp_per_weight column exists (create if needed)
if 'hp_per_weight' not in above_median_mpg.columns:
    above_median_mpg['hp_per_weight'] = above_median_mpg['horsepower'] / above_median_mpg['weight']

# Step 4: Find the max hp_per_weight ratio
max_ratio = above_median_mpg['hp_per_weight'].max()

# Step 5: Get all cars with this max ratio
best_cars = above_median_mpg[above_median_mpg['hp_per_weight'] == max_ratio]

# Display the result
print("Car(s) with best horsepower-to-weight ratio among those with above-median mpg:")
print(best_cars[['horsepower', 'weight', 'mpg', 'hp_per_weight']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Group data by model_year and origin, then calculate average mpg
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Step 2: Create the multi-line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')

# Step 3: Customize plot
plt.title('Average MPG over the Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
import seaborn as sns
import matplotlib.pyplot as plt

# Set style for cleaner visuals
sns.set(style="whitegrid")

# Reset index to avoid duplicate index issues in plotting
df_reset = df.reset_index()

# Step 1: Create the scatter plot
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    data=df_reset,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    palette='Set2',
    hue_order=['japan', 'europe', 'usa'],
    sizes=(20, 200),
    alpha=0.8,
    edgecolor='black'
)

# Step 2: Add titles and labels
plt.title('Car Weight vs Horsepower\n(Size = MPG, Color = Origin)', fontsize=14)
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
import pandas as pd

# Reload the original data to get the 'name' column back
df_with_name = pd.read_csv("Cars.csv")

# Step 1: Group by car name and calculate required stats
car_stats = df_with_name.groupby('name').agg(
    num_years=('model_year', pd.Series.nunique),
    mpg_std=('mpg', 'std'),
    avg_mpg=('mpg', 'mean')
).reset_index()

# Step 2: Filter for consistent models
consistent_cars = car_stats[
    (car_stats['num_years'] >= 2) &
    (car_stats['mpg_std'] < 1.0)
]

# Step 3: Sort by number of appearances (descending), then avg_mpg
consistent_cars_sorted = consistent_cars.sort_values(
    by=['num_years', 'avg_mpg'],
    ascending=[False, False]
)

# Step 4: Display result
print("Consistent Car Models (Low mpg variation):")
print(consistent_cars_sorted)
```
